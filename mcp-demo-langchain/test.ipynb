# Chatbot → MCP Flow

The workflow for this application is:

1. **Input**: User interacts with the chatbot.
2. **Chatbot**: The chatbot uses an LLM to decide whether a tool call is needed.
3. **MCP Protocol**: The chatbot sends the request through the MCP protocol.
4. **MCP Server**: The server invokes connected tools such as:
   - addition
   - multiplication
   - weather API
5. **Feedback**: The result is returned back to the chatbot and presented to the user.
