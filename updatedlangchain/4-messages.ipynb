# Message prompt

pass list of message objects


### message types
System, Human, AI message Tool message

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage, ToolMessage

messages = [
    SystemMessage(content="You are a helpful assistant."),
    HumanMessage(content="How to create rest API?")
]

response = model.invoke(messages)
(response.content)
```

## detailed system message

```python
system_msg = SystemMessage("""You are a coding assistant.
Expert in web developlopment. Provide code snippets and explanations to users' queries.
Be consise and clear in your responses. Always provide code examples when relevant.
"""
)
messages = [
    system_msg,
    HumanMessage(content="How to create rest API?")
]
response = model.invoke(messages)
print(response.content) 
```

## role+content+metadata

```python
human_msg = HumanMessage(
    content="Hello",
    name="user1", # for identifying the user
    id="123" # for tracking purpose
)
```

```python
response = model.invoke([human_msg])
response
```

```python
ai_msg = AIMessage("Hello! How can I assist you today? My name is AI ")
messages = [
    SystemMessage(content="You are a helpful assistant."),
    HumanMessage(content="Can you help me"),
    ai_msg,
    HumanMessage(content="What did I ask you before?")
]
response = model.invoke(messages)
print(response.content)
```

```python
response.usage_metadata
```

```python
ai_message = AIMessage(
    content=[],
    tool_calls=[
        {
            "name": "get_weather",
            "args": {"location": "Paris"},
            "id": "tool_call_123"
        }
    ]
)

weather_result = "Sunny, 25°C"
tool_message = ToolMessage(
    content=weather_result,
    tool_call_id="tool_call_123"
)

messages = [
    SystemMessage(content="You are a helpful assistant."),
    HumanMessage(content="What's the weather in Paris?"),
    ai_message,
    tool_message
]
response = model.invoke(messages)
```
