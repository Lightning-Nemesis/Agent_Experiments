# STREAMING AND BATCH

streaming for streaming response when it is getting generated for long response

```python
model.invoke("Why parrots have colorful feathers?")
```

```python
for chunk in model.stream("Why parrots have colorful feathers?"):
    print(chunk.content, end=" |",flush=True)
```

# Batch

can significantly improve performance and reduce cost, processing in parallel

```python
response = model.batch(
    [
        "Why parrots have colorful feathers?",
        "How airplaines fly?",
        "What is the capital of Italy?"
    ]
)

for res in response:
    print(res)
```

```python
model.batch(
    [
        "Why parrots have colorful feathers?",
        "How airplaines fly?",
        "What is the capital of Italy?"
    ],
    config={"temperature": 0.5,'max_concurrency':5}
)
```
